```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap
from sklearn.decomposition import PCA

import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('data.csv')
dataset
```

```python
#drop kolom Unnamed: 32 karena berisi null value dan drop kolom id karena tidak digunakan
dataset.drop(['Unnamed: 32','id'], axis = 1 , inplace=True)
dataset.head(10)
```

```python
dataset.rename(columns = {'diagnosis':'target'}, inplace = True)
dataset.head(2)
```

```python
print(dataset.target.value_counts())
```

```python
dataset['target'] = dataset['target'].replace('B', 0)
dataset['target'] = dataset['target'].replace('M', 1)
```

```python
corr_mat = dataset.corr()
corr_mat
```

```python
plt.figure(figsize=(20,10)) 
sns.heatmap(dataset.corr(), annot=True) 
plt.title("Korelasi antar fitur")
plt.show()
```

```python
#threshold untuk melihat fitur yang korelasinya paling tinggi
threshold = 0.75
filter_ = np.abs(corr_mat['target']) > threshold
corr_fitur = corr_mat.columns[filter_].tolist()
corr_fitur
```

```python
sns.heatmap(dataset[corr_fitur].corr(), annot = True, fmt = ".2f")
plt.title("Korelasi Antar Fitur")
plt.show()
```

```python
X = dataset[['concave points_mean', 'radius_worst', 'perimeter_worst', 'concave points_worst']]
y = dataset[['target']]
```

```python
#train-test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

print("Training set:", X_train.shape)
print("Test set:", X_test.shape)
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
y_train = np.array(y_train)
y_test = np.array(y_train)
```

```python
from numpy import column_stack


y_train = y_train.reset_index().drop("index", axis = 1)
X_train = pd.DataFrame(X_train, columns = column_stack)
train_dataset = pd.concat([X_train, y_train], axis = 1)
train_dataset
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion='entropy', max_depth=3)
```

```python
clf.fit(X_train, y_train)
```

```python
from sklearn.datasets import load_iris
from sklearn import tree
clf = tree.plot_tree(clf)
```

```python
score = tree.score(X_test, y_test)
print("Akurasi:", score)
```

```python
from sklearn.metrics import f1_score

# Make predictions on the test data
y_pred = clf.predict(X_test)

# Calculate the F1 score
f1 = f1_score(y_test, y_pred, average='macro')
print("F1 Score: {:.2f}".format(f1))
```
